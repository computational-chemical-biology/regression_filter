### A função `matchFeats` serve para parear dados analisados em MS1 e DDA

```python
matchFeats(samp.feat_tab.loc[49, ['row ID', 'row m/z', 'row retention time']],
           samp_006dda, ppm=10, rtabs=15)
```

```python
%%time 

qlist = []

for q in samp3.sel[samp.sel==1].index:
    qlist.append(matchFeats(samp3.feat_tab.loc[q, ['row ID', 'row m/z', 'row retention time']], samp_006dda))
```

```python
# dos 133 ions selecionados para fragmentacao
# encontramos um ms1 semelhante em dda
qtab = pd.concat(qlist).reset_index()
len(qtab['q-id'].unique())
```

```python
%%time 

qlist2 = []

for q in samp3.sel[samp.sel==1].index:
    qlist2.append(matchFeats(samp3.feat_tab.loc[q, ['row ID', 'row m/z', 'row retention time']], samp_006ddams1))
```
